```python
#count can be used instead of size
desig_data=data.groupby('Designation')['Attrition '].size().reset_index(name='count')
desig_data
```

```python
data.groupby('Tenure').count()
```

```python
data.Tenure=data.Tenure.str.replace(' ','')
```

```python
large=22
med=16
small=12
```

```python
%matplotlib inline
```

```python
params = { 'axes.titlesize': large, 'legend.fontsize' : med, 'xtick.labelsize' : small , 'ytick.labelsize' : small, 'axes.labelsize' : med , 'figure.figsize' : (10,10)  }
```

```python
plt.rcParams.update(params)
```

```python
fig, ax = plt.subplots()

# explode attribute will offset ,i.e, pull out the corresponding wedge by the mentioned fraction of radii.
#autopct will bd displayed on the pie. pct means percentage

wedges, text, autotexts = ax.pie(desig_data['count'],labels=desig_data['Designation'], autopct=lambda pct: pct,textprops=dict(color='w') )

# wedges, text is used to gulp in unnecessary data which would otherwise be printed on screen

ax.legend(wedges, (desig_data['Designation']),title='Designation',loc='lower right',bbox_to_anchor=(0,0,1.5,0))

ax.set_title('Attrition based on Designation')

```

```python
plt.bar(desig_data['Designation'],desig_data['count'],color=('b','y','g','r','b','y','g'))
plt.gca().set_xticklabels(desig_data['Designation'],rotation=90)
plt.title('Attrition based on Designation')
plt.ylabel("Number of Employees")
plt.xlabel("Designation")

#This is an easy way of plotting without bothering about the underlying figure, subplots and axes.

# Note : for plt it is title but for axes it is set_title
```

```python
sns.boxplot(x='count', y='Designation', data=desig_data['count'], order=desig_data['Designation'])
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1)
sns.kdeplot(data['Tenure'],data['Age'],shade=True,ax=ax[0])
ax[0].set_title('Bivariate Distribution Function')

sns.kdeplot(data['Tenure'],color='r',shade=True,ax=ax[1])
sns.kdeplot(data['Age'],color='g',shade=True,ax=ax[1])
ax[1].set_title('Univariate Distribution Function')

#For sns the axes needs to be mentioned inside as an argument. For others like pie, bar, we would do axes.pie,.... 
```

```python
plt.locator_params(axis='y',nbins=10)
sns.stripplot(data['Age'],data['Tenure'])
```

```python
sns.boxplot(x=data['Gender'],y=data['Monthly Income'])
plt.title('Boxplot Salary vs. Gender')
plt.ylim(0,50000)

# x and y can be interchanged to get horizontal boxplots

# Boxplot indicates quartiles. Q1,Q2,Q3. Central line in coloured region is Q2 i.e. median. Horizontal Bars at end of tails 
#is max/min non outliers
# Outliers indicated by dots
```

```python
sns.violinplot(x=data.Gender,y=data['Monthly Income'])
plt.title('Violin Plot for Salary w.r.t Gender')

#Violin Plot is a combination of Box Plot and Density Plot. White Circle is the Median and 
#the black region is the interquartile region
```
